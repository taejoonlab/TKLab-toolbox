```python
def read_count(filename):
    rv = dict()
    f = open(filename,'r')
    f.readline()
    for line in f:
        tokens = line.strip().split("\t")
        rv[ tokens[0].split('|')[0] ] = tokens[1:]
    f.close()
    return rv

p_count = read_count('../../data/Flounder/Kwon201608_KoreanFlounder_Liver.count.txt')
#d_count = read_count('../../data/Flounder/Yang2012_DANREtx_liver.count.txt')
d_count = read_count('../../data/Flounder/TissueMap201711_MOUSE_Liver.count.txt')

f_out = open('../../data/Flounder/PALOR-MOUSE_liver.count.txt','w')
f_pair = open('../../data/Flounder/PALOR-MOUSE.best_pairs','r')
for line in f_pair:
    tokens = line.strip().split("\t")
    p_name = tokens[0].split('|')[0]
    d_name = tokens[1].split('|')[0]
    
    d_count_str = "0\t0\t0"
    if d_name in d_count:
        d_count_str = '\t'.join(d_count[d_name])
    p_count_str = "0\t0\t0"
    if p_name in p_count:
        p_count_str = '\t'.join(p_count[p_name])
        
    f_out.write("%s|%s\t%s\t%s\n"%(d_name, p_name, p_count_str, d_count_str))
f_pair.close()
f_out.close()
```
