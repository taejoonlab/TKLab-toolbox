```python
import os
import sys

filename_m5 = '../../data/degRNAprobe/HUMAN_ens87_cdna_kmer.20.multi5'

#for tmp_f in os.listdir('../../data'):
#    print(tmp_f)
#sys.exit(1)

p2freq = dict()
p2t = dict()
f_m5 = open(filename_m5,'r')
for line in f_m5:
    tokens = line.strip().split("\t")
    tmp_seq = tokens[0]
    tmp_freq = int(tokens[1])
    if tmp_freq > 500:
        continue
        
    t_list = ['.'.join(x.split('.')[:-1]) for x in tokens[2].split(',')]
    p2freq[tmp_seq] = tmp_freq
    p2t[tmp_seq] = t_list
f_m5.close()

cum_t_list = []
tmp_count = 0
prev_t_count = 0
for tmp_p in sorted(p2freq.keys(), key=p2freq.get, reverse=False):
    if len(p2t[tmp_p]) < 100:
        continue
        
    cum_t_list = list(set(cum_t_list + p2t[tmp_p]))
    tmp_t_count = len(cum_t_list)
    if tmp_t_count == prev_t_count:
        continue

    tmp_count += 1    
    #print(tmp_p, tmp_count/len(p2freq), len(p2t[tmp_p]), len(cum_t_list))
    #if tmp_count > 5000:
    #    break
    
print( tmp_count, len(cum_t_list) )
```
